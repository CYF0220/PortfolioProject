```python
X = df.drop('price',axis = 1).values
y = df['price'].values
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import log_loss
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=71)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train= scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
from tensorflow.keras.optimizers import Adam
```

```python
#scores = []
kf = KFold(n_splits= 4,shuffle=True,random_state = 71)
for tr_idx, va_idx, in kf.split(X_train):
    tr_x, va_x = X_train[tr_idx], X_train[va_idx]
    tr_y, va_y = y_train[tr_idx], y_train[va_idx]
    
    model = Sequential()
    model.add(Dense(19,activation = 'relu'))
    
    model.add(Dense(19,activation = 'relu'))
    model.add(Dropout(0.2))
    
    model.add(Dense(19,activation = 'relu'))
    model.add(Dropout(0.2))
    
    model.add(Dense(1))

    model.compile(optimizer='adam',loss='mse')
    
    model.fit(x=tr_x,y=tr_y,
          validation_data=(va_x,va_y),
          batch_size=128,epochs=400)
    losses = pd.DataFrame(model.history.history)
    losses.plot()
    
#    va_pred = model.predict(va_x)
#    score = mean_squared_error(va_y,va_pred)
#    scores.append(score)

#print('mse:', np.mean(scores))    
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
mean_absolute_error(y_test,predictions)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
explained_variance_score(y_test,predictions)
```
